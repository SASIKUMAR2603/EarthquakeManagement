1.Compare any colomn with 4 or 5

```python
country_counts=ds['country'].value_counts().reset_index()
country_counts.columns=['country','earthquake_count']
country_counts.head(20)
```

2.add new column

```python
selected_countries=['Indonesia','Japan','India','Solomon Islands']
filtered_country_counts=country_counts[country_counts['country'].isin(selected_countries)]
plt.figure(figsize=(10,6))
sns.barplot(x='country',y='earthquake_count',data=filtered_country_counts)
plt.title('Earthquake Count for Selected Countries')
plt.xlabel('Country')
plt.ylabel('Earthquake Count')
plt.xticks(rotation=45)
plt.show()
```

```python
ds.hist(figsize=(10,10),bins=1,color='red')
plt.title('Histogram of Earthquake data', fontsize=20, color='Blue')
plt.show()
```

1.Show Boxplot

```python
sns.boxplot(x=ds['depth'])
plt.title('Boxplot of Depth',fontsize=20,color='r')
plt.xlabel('Depth')
plt.ylabel('Frequency')
plt.show()
```

2.Show Barplot

```python
sns.barplot(x=ds['country'],y=ds['depth'])
plt.title('Barplot of Depth',fontsize=15,color='black')
plt.xlabel('Country')
plt.ylabel('Magnitude')
plt.xticks(rotation=45)
plt.show()
```

3.Show Scatterplot

```python
sns.scatterplot(x=ds['country'],y=ds['depth'],data=ds,palette='rocket',color='y')
plt.title('Scatterplot of Depth',fontsize=25,color='Green')
plt.xlabel('Country')
plt.ylabel('Depth')
plt.xticks(rotation=60)
plt.show()
```

4.Show Heatmap

```python
numerical_cols = ds.select_dtypes(include=np.number)
correlation_matrix = ds[numerical_cols.columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
